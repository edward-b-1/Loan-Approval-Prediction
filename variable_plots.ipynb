```python
import pandas
import numpy
import matplotlib.pyplot as plt
import itertools
import sklearn
from sklearn.metrics import log_loss
from sklearn import preprocessing
```

```python
import loan_approval_lib
from loan_approval_lib import logspace, linspace
```

```python
data = loan_approval_lib.load_original_data()
```

```python
data_loan_status_deny = data[data['loan_status'] == 1]
data_loan_status_approve = data[data['loan_status'] == 0]
```

```python
print(len(data_loan_status_deny), len(data_loan_status_approve))
```

```python
column = 'person_age'

data_accept = data_loan_status_approve[column].to_numpy()
data_reject = data_loan_status_deny[column].to_numpy()

_min = min(data_accept.min(), data_reject.min())
_max = max(data_accept.max(), data_accept.min())
print(_min, _max) # both even so ok
_steps = int(_max - _min) // 2
bins = numpy.linspace(_min, _max, _steps)

plt.hist(
    data_accept,
    bins=bins,
    histtype='step',
    density=True,
    label='approve',
    linewidth=2,
)
plt.hist(
    data_reject,
    bins=bins,
    histtype='step',
    density=True,
    label='reject',
    linewidth=2,
)
plt.ylabel('density')
plt.xlabel(column)
plt.legend()
plt.yscale('log')
plt.xlim(xmin=0)
plt.savefig(f'{column}_explore.pdf')
None
```
