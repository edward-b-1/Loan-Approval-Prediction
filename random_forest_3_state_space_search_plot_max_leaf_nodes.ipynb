# Plot

```python
import math
import pandas
import numpy
import itertools
import matplotlib.pyplot as plt
```

```python
filename = 'random_forest_3_state_space_search_max_leaf_nodes.csv'
```

```python
df = pandas.read_csv(filename, index_col=0)
```

```python
df
```

```python
plt.figure()

positions = []
x = []
column = 'max_leaf_nodes'
metric = 'accuracy_in_sample'
x_label = 'maximum number of leaf nodes'
y_label = 'in sample accuracy'
box_width = 0.1
x_min = 0.9
x_max = 1100

for value in sorted(df[column].unique()):
    x.append(df[df[column] == value][metric])
    positions.append(value)

plt.boxplot(
    x,
    positions=positions,
    orientation='vertical',
    meanline=True,
    showmeans=True,
    widths=[box_width * position for position in positions],
)
plt.xscale('log')
plt.xlim(xmin=x_min, xmax=x_max)
plt.xlabel(x_label)
plt.ylabel(y_label)

None
```
